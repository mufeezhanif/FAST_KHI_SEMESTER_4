```python
df = pd.read_csv('heart_disease.csv')
df
```

```python
df.shape #10000 rows 21 cols
```

```python
df.describe()
```

```python
df[['Gender', 'Smoking']].nunique()
```

```python
df.isnull().sum()
```

```python
numColumns = df.select_dtypes(include=['float64']).columns

# Handling missing values for both numerical & categorical columns
for col in numColumns:
    if df[col].isnull().sum() > 0:  
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr

        outliers = df[(df[col] < lower) | (df[col] > upper)]

        if not outliers.empty:
            df[col].fillna(df[col].median(), inplace = True)
        else:
            df[col].fillna(df[col].mean(), inplace = True)

df.isnull().sum()
```

```python
qualitativeColumns = df.select_dtypes(include=['object']).columns

for col in qualitativeColumns:
    df[col].fillna(df[col].mode().iloc[0], inplace=True)
        

df.isnull().sum()
```

```python
plt.figure(figsize=(15, 10))
sns.histplot(df['Age'], bins=9)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(x=df["Heart Disease Status"])
```
